# CREATING WORKFLOW

```python
import requests
import pandas as pd
from pydoc import importfile
module = importfile('..\\config.py')
from config import RAPID_API_KEY
API_HOST = "flixster.p.rapidapi.com"
POPULAR_MOVIES_URL = "https://flixster.p.rapidapi.com/movies/get-popularity"
MOVIE_DETAIL_URL = "https://flixster.p.rapidapi.com/movies/detail"
HEADERS = {
	"X-RapidAPI-Key": RAPID_API_KEY,
	"X-RapidAPI-Host": API_HOST
}
```

```python
# get movie details for given movie id 
# Requires api key for making the api call
# For null values returns None
def get_movie_details(movie_id, api_key):
    """Make a API call for the given movie's details. Returns some info from the API response """
    headers = {
	"X-RapidAPI-Key": api_key,
	"X-RapidAPI-Host": API_HOST
}
    querystring = {"emsVersionId":movie_id}

    response_details = requests.get(MOVIE_DETAIL_URL, headers=headers, params=querystring)
    data_details = response_details.json()
    try:
        movie_duration_time = (data_details['data']['movie']['durationMinutes'])
    except:
        movie_duration_time = 'None'
    try:
        movie_release_date = data_details['data']['movie']['releaseDate']
    except:
        movie_release_date='None'
    try:
        movie_total_gross = data_details['data']['movie']['totalGross']
    except:
         movie_total_gross = 'None'
    try:
        movie_director = data_details['data']['movie']['directedBy']
    except:
        movie_director = 'None'
    try:
        movie_synopsis = data_details['data']['movie']['synopsis']
    except:
        movie_synopsis = 'None'
    try:
        movie_main_genre = data_details['data']['movie']['genres'][0]['name']
    except:
        movie_main_genre = 'None'

    return movie_director, movie_release_date, movie_synopsis, movie_total_gross, movie_duration_time, movie_main_genre
```

```python
# get popular movie and its details as a dataframe
# Make API call with given api key for getting popular movies
# then for every popular movie make an API call for its detail
# then add the data as a row to the dataframe
# Return the dataframe 

def get_popular_movie_data(api_key):
    headers = {
	"X-RapidAPI-Key": api_key,
	"X-RapidAPI-Host": API_HOST
}
    response_popular = requests.get(POPULAR_MOVIES_URL, headers=headers)
    data_popular =response_popular.json()
    df= pd.DataFrame(columns = [
                "movie_id","movie_ver_id","movie_title",
                "movie_director","movie_release_date","movie_synopsis",
                "movie_duration_time","movie_main_genre",
                "movie_total_gross","movie_rating"])

    for movie in data_popular['data']['popularity']:
        movie_id = movie['emsId']
        movie_ver_id = movie['emsVersionId']
        movie_title = movie['name']
        try:
            movie_rating = movie['userRating']['dtlLikedScore']
        except:
            movie_rating='None'
        
        movie_details =get_movie_details(api_key=api_key,movie_id=movie_ver_id)
        movie_director, movie_release_date, movie_synopsis,movie_total_gross,movie_duration_time, movie_main_genre = movie_details
        df = df._append({
                "movie_id": movie_id,
                "movie_ver_id": movie_ver_id,
                "movie_title": movie_title,
                "movie_director": movie_director,
                "movie_release_date": movie_release_date,
                "movie_synopsis": movie_synopsis,
                "movie_duration_time": movie_duration_time,
                "movie_main_genre": movie_main_genre,
                "movie_total_gross": movie_total_gross,
                "movie_rating": movie_rating,
             
            }, 
            ignore_index = True
            )
    return df
    
    
```

```python
popular_movies_df = get_popular_movie_data(api_key=RAPID_API_KEY)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Movie's rating usually between approximately 80-100
plt.figure(figsize=(15,8))
sns.histplot(data=popular_movies_df, x='movie_rating', bins=5,  kde=True, color='green' )
```

```python
popular_movies_df['movie_total_gross'].max()
```

```python
# Showing the movies that has more than 10^6 gross
plt.title('Popular Movies Total Gross')
plt.ylabel('Total Gross (hundred million)')
plt.xlabel('Movies')
plt.tick_params(axis='both', which='both', labelbottom=False, bottom=False)

for i in range(len(popular_movies_df)):
    try:
        x = popular_movies_df['movie_title'][i]
        y = popular_movies_df['movie_total_gross'][i]
        plt.plot(x, y, "o", color="hotpink")
        if popular_movies_df['movie_total_gross'][i] > 100000000:
            plt.text(x, y * (1 - 0.05), popular_movies_df['movie_title'][i], fontsize=10)
    except:
        continue
plt.show()

```

```python
# Only popular movie that doesn't release in 2023 is die hard and love actually 
plt.title('Popular Movies Release years')
plt.ylabel('Release year')
plt.xlabel('Movies')
plt.tick_params(axis='both', which='both', labelbottom=False, bottom=False)

for i in range(len(popular_movies_df)):
    try:
        x = popular_movies_df['movie_title'][i]
        y = popular_movies_df['movie_release_date'][i]
        plt.plot(x, y, "o", color="coral")
        if popular_movies_df['movie_release_date'][i].year < 2023:
            plt.text(x, y, popular_movies_df['movie_title'][i], fontsize=10)
    except:
        continue

        
plt.show()
```
