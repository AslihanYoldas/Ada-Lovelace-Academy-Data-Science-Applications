```python
#Checking the outlayers to be sure
#Calculating z-score finding outliers and returning them
def find_outliers(column_name):
    threshold = 3
    outliers = []
    for index, data in enumerate(sleep_df[column_name]):
        z_score = (data - sleep_df[column_name].mean()) / sleep_df[column_name].std()
        if abs(z_score) > threshold:
            outliers.append(index+1)
    return outliers
```

```python
outliers=find_outliers('Caffeine_consumption')
```

### Corelation Matris

```python
# Calculating corelation
# Deep_sleep_percentage and light_sleep_percentage has negative corelation
# Deep_sleep_percentage and sleep efficency has positive corelation

pd.set_option('display.precision', 5)

matrix = sleep_df[['Age', 
    'Sleep_duration',
    'Sleep_efficiency',
    'REM_sleep_percentage',
    'Deep_sleep_percentage',
    'Light_sleep_percentage', 
    'Awakenings', 
    'Caffeine_consumption',
    'Alcohol_consumption', 
    'Exercise_frequency'
    ]].corr()
matrix
```

### Heatmap

```python
# We can see corelation more clearly here
# light colors (more like white) means positive corelation
# dark colors (more like purple) means negative corelation
# more like pink colors mean no corellation
# We can also see here
# Deep_sleep_percentage and light_sleep_percentage has negative corelation
# sleep efficency and light_sleep_percentage has negative corelation
# Deep_sleep_percentage and sleep efficency has positive corelation
sns.heatmap(matrix, 
            xticklabels=matrix.columns.values,
            yticklabels=matrix.columns.values)
```
