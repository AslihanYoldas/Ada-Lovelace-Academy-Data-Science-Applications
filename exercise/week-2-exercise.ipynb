```python
# Merging dataframes of Turkey's data
Turkey_covid = pd.concat([infected_data_date_index['Turkey'],deaths_data_date_index['Turkey'], recovered_data_date_index['Turkey']], axis=1, join="inner")
```

```python
Turkey_covid.columns = ['infected' , 'deaths', 'recovered']
```

```python
#Getting the data where death numbers more than recovery numbers in Turkey
Turkey_covid[Turkey_covid['deaths']>Turkey_covid['recovered']]
```

```python
# Getting total number of infections, recoverys, deaths  in Turkey
Turkey_covid.sum()
```

```python
#Getting the avarege of infections, recoverys, deaths in Turkey
Turkey_covid.mean()
```

```python
#Standartization using apply
Turkey_covid['infected'].apply(lambda x : (x-Turkey_covid['infected'].mean())/Turkey_covid['infected'].std()).to_frame()
```
