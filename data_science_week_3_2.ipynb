```python
import pandas as pd
#from sklearn.datasets import load_iris()
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as mcolors
import seaborn as sns
```

```python
df = pd.DataFrame(
    {'Group': ['A', 'B', 'C', 'D', 'E' ],
     'Value': [1, 5, 4, 3, 9]

    }
)
```

```python
plt.barh(y=df.Group, width=df.Value)
```

```python
df = df.sort_values(by = ['Value'])
```

```python
# Amaç yorumlamayı doğru yapabilmek , göze daha anlaşılır gelmesi, anlamlandırmayı kolaylaştırmak
plt.barh(y=df.Group,
          width=df.Value,
          color =['red', 'blue', 'green'],
          edgecolor='black'
          )
```

```python
my_plot = plt.barh(y=df.Group,
          width=df.Value,
          color =mcolors.TABLEAU_COLORS,
          edgecolor='black',
          linewidth=2
          )
plt.bar_label(container= my_plot, padding=3)
plt.show()
```

```python
tips= sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.set(style = 'darkgrid')
sns.barplot(
    x = 'total_bill',
    y = 'day',
    data = tips,
    color = 'cyan'
)
plt.figure(figsize = (20, 10))
```

```python
g = sns.FacetGrid(tips, col="day", height=4, aspect=.5)
g.map(sns.barplot, "sex", "total_bill", order=["Male", "Female"])
```

```python
sns.barplot(
        x="total_bill",
        y="smoker",
        hue="sex",
        data=tips,
        color="pink"
        )
plt.figure(figsize=(16, 8))
```

```python
total = tips.groupby(['day'])['total_bill'].sum().reset_index()
```

```python
total
```

```python
bar1 = sns.barplot(x='day', y='total_bill', data=total)
```

```python
smoker = tips[tips.smoker=='Yes']
```

```python
bar2 = sns.barplot(
    x ='day',
    y='total_bill',
    data=smoker,
    estimator =sum,
    errorbar=None,
)

```

```python
from matplotlib import patches
bar1 = sns.barplot(x='day', y='total_bill', data=total, color= 'darkblue')
bar2 = sns.barplot(
    x ='day',
    y='total_bill',
    data=smoker,
    estimator =sum,
    errorbar=None,
)
top_bar = patches.Patch(color = 'darkblue', label = 'non-smoker')
bottom_bar = patches.Patch(color = 'lightblue', label = 'smoker')
plt.legend(handles=[top_bar, bottom_bar])
plt.show()
```

```python
df_iris = sns.load_dataset('iris')
```

```python
df_iris.head()
```

```python
sns.histplot(data=df_iris, x='sepal_length',  kde=True)
```

```python
sns.displot(df_iris["sepal_length"])
#displot is deprecated
```

```python
#Çok kümelenmiş verilerde anlamlı olmuyor. Üst üste biniyor
sns.regplot(x=df_iris['sepal_length'],
             y= df_iris['sepal_width'], 
             marker='x',
             line_kws ={
                 "color" :"red",
                 "alpha" : 0.6,
                 "lw":5
             },
             scatter_kws={
                 "color":"darkblue",
                 "s":20
             }
             )
```

```python
sns.lmplot(
        x='sepal_length',
        y= 'sepal_width', 
        data=df_iris,
        hue="species",
        fit_reg=False
          
     )
```

```python
#boxplotoutlayer mean quartile verinin dağılımıyla ilgili yorumlar yapabiliyoruz
sns.set_theme(style="darkgrid")
sns.boxplot(x=df_iris['species'], y=df_iris['sepal_length'])
```

```python
df_iris.boxplot(column='sepal_length', by='species')
plt.show()
```
